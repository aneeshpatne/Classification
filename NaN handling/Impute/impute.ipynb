```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Datasets/data_gov.csv")
df.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
numeric_df.head()
```

```python
imputer = SimpleImputer(strategy='median')
new_values = imputer.fit_transform(numeric_df)
new_values
```

```python
new_df = pd.DataFrame(new_values , columns = numeric_df.columns)
new_df.head()
```

```python
imputer = SimpleImputer(strategy='mean')
new_values = imputer.fit_transform(numeric_df)
new_values
```

```python
new_df = pd.DataFrame(new_values , columns = numeric_df.columns)
new_df.head()
```
